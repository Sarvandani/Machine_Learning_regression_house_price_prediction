```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Generate area data (sq ft) in a reasonable range
np.random.seed(0)  # For reproducibility
area = np.array([500, 600, 700, 800, 900, 1000, 1100, 1200, 1300, 1400, 1500,
                 1600, 1700, 1800, 1900, 2000, 2100, 2200, 2300, 2400, 2500,
                 2600, 2700, 2800, 2900, 3000, 3100, 3200, 3300, 3400, 3500])

# Generate approximate prices with some noise for realism
actual_price = 0.2 + 0.00015 * area + np.random.normal(0, 0.02, size=area.size)

# Reshape area data for sklearn model
area_reshaped = area.reshape(-1, 1)

# Linear regression model
model = LinearRegression()
model.fit(area_reshaped, actual_price)
predicted_price = model.predict(area_reshaped)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(area, actual_price, color="blue", label="Actual Price", s=50)
plt.plot(area, predicted_price, color="red", label="Predicted Price", linewidth=2)
plt.xlabel("Area (sq ft)")
plt.ylabel("Price ( Million)")
plt.title("Actual vs Predicted Price based on Area")
plt.legend()
plt.grid(True)
plt.show()
```
